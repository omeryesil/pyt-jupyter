````python
from pandas_datareader import data as web # same to: import data as wb

MSFT = web.DataReader("MSFT", data_source="yahoo", start="2019-01-01", end="2020-1-1")
MSFT
````

````python
## Verifying data 
MSFT.info()
````

````python
## First 5 rows 
MSFT.head()
````

````python
## With X number of rows from the top
MSFT.head(3) # last x rows are the same: MSFT.tail(13)
````

````python
## Getting multiple stocks, and only "Adj Close" column
tickers = ['PG', 'MSFT', 'T', 'F', 'GE']
adjustedPrices = pd.DataFrame()
for t in tickers:
    adjustedPrices[t] = web.DataReader(t, data_source="yahoo", start="2019-01-01")["Adj Close"]
adjustedPrices.tail()    
    
````
