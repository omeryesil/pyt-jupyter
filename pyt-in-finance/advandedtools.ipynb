## Quandl

**Reference:** https://www.quandl.com/tools/python 

Quandl unifies financial and economic datasets from hundreds of publishers on a single user-friendly platform.

**To install**
```bash
conda install -c anaconda quandl
```

Getting financial data is amazingly easy with the Quandl Python module. For example, to get US GDP from FRED, just do this:

```python
import quandl
mydata = quandl.get("FRED/GDP")
```

Wothout registration, we can make maxiumum 50 calls per day, if you want to make more calls you need to registeron quandl's web site, which is free. Once we register, we get an API key that can be used in the code like:

```python
quandl.ApiConfig.api_key = "<Your_API_Key>" 
```


````python
import quandl
import os 

quandl.ApiConfig.api_key = os.environ.get("QUANDL_API_KEY") # getting api key from environment variable  

mydata = quandl.get("FRED/GDP", start_date="1950-01-01", end_date="1955-12-31")
mydata.head()
````

````python
# Saving results to file -------------
folder = "~/Documents"

# saving to csv 
csvFilePath = "{}/fred_gdp_1950-1955.csv".format(folder)
mydata.to_csv(csvFilePath) 

#saving to excel
xlsFilePath = "{}/fred_gdp_1950-1955.xlsx".format(folder)
mydata.to_excel(xlsFilePath) 
````

````python
# Reading from files ------------------
import pandas as pd

mydata_fromCsv = pd.read_csv(csvFilePath)
mydata_fromXsl = pd.read_excel(xlsFilePath)


# Time Series ---------------------------------
# here, python automcatically adds index
print (mydata_fromCsv)

# we can add set a custom index after the load
mydata_fromCsv.set_index('Date',inplace=True, drop=True) # we need to drop the existing one
print(mydata_fromCsv)

# OR here we use a custom column for the index while loading
mydata_fromCsv = pd.read_csv(csvFilePath, index_col="Date")
print(mydata_fromCsv)



````
